```python
# ===============================================
# 0. IMPORTS
# ===============================================
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, ConfusionMatrixDisplay
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
import matplotlib.pyplot as plt

# ===============================================
# 1. LOAD DATA
# (Upload your CSVs to Colab first)
# ===============================================

scaled = pd.read_csv("horse_scaled.csv")          # scaled, encoded
unscaled = pd.read_csv("horse_No_scaling.csv")    # encoded, not scaled

# Outcome column name
label_col = "outcome"

# ===============================================
# 2. TRAIN/TEST SPLITS
# ===============================================

# For scaled dataset (LogReg, LinearSVM)
X_scaled = scaled.drop(columns=[label_col])
y_scaled = scaled[label_col]

Xs_train, Xs_test, ys_train, ys_test = train_test_split(
    X_scaled, y_scaled, test_size=0.2, random_state=42, stratify=y_scaled
)

# For unscaled dataset (RF, GBM)
X_unscaled = unscaled.drop(columns=[label_col])
y_unscaled = unscaled[label_col]

Xu_train, Xu_test, yu_train, yu_test = train_test_split(
    X_unscaled, y_unscaled, test_size=0.2, random_state=42, stratify=y_unscaled
)

# ===============================================
# 3. MODELS
# ===============================================

results = {}

# LOGISTIC REGRESSION (scaled)
logreg = LogisticRegression(max_iter=2000)
logreg.fit(Xs_train, ys_train)
pred_lr = logreg.predict(Xs_test)
results["Logistic Regression"] = (ys_test, pred_lr)

# LINEAR SVM (scaled)
svm = SVC(kernel="linear", probability=True)
svm.fit(Xs_train, ys_train)
pred_svm = svm.predict(Xs_test)
results["Linear SVM"] = (ys_test, pred_svm)

# RANDOM FOREST (unscaled)
rf = RandomForestClassifier(n_estimators=300, random_state=42)
rf.fit(Xu_train, yu_train)
pred_rf = rf.predict(Xu_test)
results["Random Forest"] = (yu_test, pred_rf)

# GRADIENT BOOSTING (unscaled)
gb = GradientBoostingClassifier(random_state=42)
gb.fit(Xu_train, yu_train)
pred_gb = gb.predict(Xu_test)
results["Gradient Boosting"] = (yu_test, pred_gb)

# ===============================================
# 4. OUTPUT: ACCURACY + REPORT + CONFUSION MATRIX
# ===============================================

for model_name, (y_true, y_pred) in results.items():
    print("\n==============================")
    print(model_name)
    print("==============================")
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print(classification_report(y_true, y_pred))

    ConfusionMatrixDisplay.from_predictions(y_true, y_pred)
    plt.title(model_name)
    plt.show()
```

```python
# ===============================================
# 0. IMPORTS
# ===============================================
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, ConfusionMatrixDisplay
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
import matplotlib.pyplot as plt

# ===============================================
# 1. LOAD DATA
# ===============================================

scaled = pd.read_csv("horse_scaled.csv")          # scaled numeric
unscaled = pd.read_csv("horse_No_scaling.csv")    # numeric, unscaled

label_col = "outcome"

# ===============================================
# 2. MAKE OUTCOME BINARY
# 0 = died
# 1 = euthanized
# 2 = lived
#
# New binary:
# 0 (died + euthanized)
# 1 (lived)
# ===============================================

def convert_to_binary(df):
    df = df.copy()
    df[label_col] = df[label_col].replace({
        0: 0,   # died  0
        1: 0,   # euthanized  0
        2: 1    # lived  1
    })
    return df

scaled_bin = convert_to_binary(scaled)
unscaled_bin = convert_to_binary(unscaled)

# ===============================================
# 3. TRAIN/TEST SPLITS
# ===============================================

# SCALED (for LogReg + Linear SVM)
X_scaled = scaled_bin.drop(columns=[label_col])
y_scaled = scaled_bin[label_col]

Xs_train, Xs_test, ys_train, ys_test = train_test_split(
    X_scaled, y_scaled, test_size=0.2, random_state=42, stratify=y_scaled
)

# UNSCALED (for RF + GBM)
X_unscaled = unscaled_bin.drop(columns=[label_col])
y_unscaled = unscaled_bin[label_col]

Xu_train, Xu_test, yu_train, yu_test = train_test_split(
    X_unscaled, y_unscaled, test_size=0.2, random_state=42, stratify=y_unscaled
)

# ===============================================
# 4. MODELS
# ===============================================

results = {}

# LOGISTIC REGRESSION (scaled)
logreg = LogisticRegression(max_iter=2000)
logreg.fit(Xs_train, ys_train)
pred_lr = logreg.predict(Xs_test)
results["Logistic Regression"] = (ys_test, pred_lr)

# LINEAR SVM (scaled)
svm = SVC(kernel="linear", probability=True)
svm.fit(Xs_train, ys_train)
pred_svm = svm.predict(Xs_test)
results["Linear SVM"] = (ys_test, pred_svm)

# RANDOM FOREST (unscaled)
rf = RandomForestClassifier(n_estimators=300, random_state=42)
rf.fit(Xu_train, yu_train)
pred_rf = rf.predict(Xu_test)
results["Random Forest"] = (yu_test, pred_rf)

# GRADIENT BOOSTING (unscaled)
gb = GradientBoostingClassifier(random_state=42)
gb.fit(Xu_train, yu_train)
pred_gb = gb.predict(Xu_test)
results["Gradient Boosting"] = (yu_test, pred_gb)

# ===============================================
# 5. REPORTS + CONFUSION MATRICES
# ===============================================

for model_name, (y_true, y_pred) in results.items():
    print("\n==============================")
    print(model_name)
    print("==============================")
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print(classification_report(y_true, y_pred, digits=4))

    ConfusionMatrixDisplay.from_predictions(y_true, y_pred)
    plt.title(model_name)
    plt.show()
```
